### 1. Escriba un programa que pida dos números enteros y que calcule su división, escribiendo si la división es exacta o no.

```python
num1 = int(input('enter the first numer'))
num2 = int(input('enter the second number'))

result = round(num1/num2, 2)

if num1%num2 == 0:
    print('the result is:  ' +str(result)+ ' is an exact division')
else:
    print('the result is:  ' +str(result)+ ' is not an exact division')
```
